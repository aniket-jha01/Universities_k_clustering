```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lets use some artificial data using scikit learn
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0]
```

```python
data[1]
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_  #gives out the centers of clusters
```

```python
kmeans.labels_  #returns the labels it believes to be true for the clusters
```

```python
#Now like the previous methods we dont really predict anything here, cuz the data is obviously not labelled
#We have the original labels , lets see we were able to find clusters or not

#Lets see how our kmeans compares to our original
#Using tuple unpacking
fig, (ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```
